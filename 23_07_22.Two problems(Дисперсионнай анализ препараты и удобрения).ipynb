```python
import numpy as np
import pandas as pd
import scipy.stats as stats 
import matplotlib.pyplot as plt 
import seaborn as sns
```

Задача 1.

```python
np.random.seed(11)
A = stats.norm.rvs(scale=30, loc=11, size=100)
B = A + stats.norm.rvs(scale=54, loc=11, size=100)

y = []
for id in range(len(A)):
  y.append([A[id], B[id]])
y = np.array(y)
data = pd.DataFrame(data=y, columns = ['Препарат A', 'Препарат B'])
```

```python
ax = sns.boxplot(data=data, palette="Blues")
plt.show()
```

```python
data['Препарат A'].hist(bins = 25, alpha = 0.5)
data['Препарат B'].hist(bins = 25, alpha = 0.5)
plt.show()
```

```python
from scipy.stats import mannwhitneyu
stat, p = mannwhitneyu(A, B) 
print('p = %.3f' %p)

alpha = 0.05
if p > alpha:
    print('Одинаковые распределения (не отвергаем гипотезу об отсутвии различий в влиянии препарата)')
else:
    print('Разные распределения (отвергаем гипотезу об отсутвии различий в влиянии препарата)')
```

**Задача 2.**

```python
water = [1,2,3,4,2,4,2,4,5,2,3,4,2,1,3,4,3,2,5,1]
nutri = [1,2,4,6,5,6,7,5,4,5,6,7,4,3,5,5,6,5,4,3,5]
mineral =[2,1,1,3,2,4,2,4,5,4,3,2,3,2,3,1,3,4,5,1,4]

y = []
for id in range(len(water)):
  y.append([water[id], nutri[id], mineral[id]])
data = pd.DataFrame(data=np.array(y), columns = ['water', 'nutri', 'mineral'])
```

```python
ax = sns.boxplot(data=data, palette="Blues")
plt.show()
```

```python
data.water.hist(bins = 7, alpha = 0.5)
data.nutri.hist(bins = 7, alpha = 0.5)
data.mineral.hist(bins = 7, alpha = 0.5)
plt.show()
```

```python
from scipy.stats import f_oneway
F, p = f_oneway(water, nutri, mineral) 
alpha = 0.05 

if p > alpha:
    print('Одинаковое распределение (не отвергаем гипотезу об отсутствии влияния типа жидкости для полива на качесчтво растений)')
else:
    print('Разное распределение (принимаем гипотезу о влиянии типа жидкости для полива на качесчтво растений)')
```
